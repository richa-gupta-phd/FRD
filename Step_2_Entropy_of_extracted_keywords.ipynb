```python
import pandas as pd
import math

def calculate_entropy_from_csv_pandas(csv_file_path, word_list):
    # Load the CSV file
    df = pd.read_csv(csv_file_path, usecols=[1])  # Assuming text is in the second column
    sentences = df.iloc[:,0].str.lower().tolist()

    # Initialize a dictionary to hold entropy values for each word
    entropy_dict = {}

    # Calculate entropy for each word in the word list
    for word in word_list:
        word = word.lower()
        # Count occurrences and non-occurrences in sentences
        occurrences = sum(word in sentence for sentence in sentences)
        non_occurrences = len(sentences) - occurrences
        # Calculate probabilities
        p_occurrence = occurrences / len(sentences) if len(sentences) > 0 else 0
        p_non_occurrence = non_occurrences / len(sentences) if len(sentences) > 0 else 0
        # Calculate entropy
        entropy = 0
        if p_occurrence > 0:
            entropy -= p_occurrence * math.log2(p_occurrence)
        if p_non_occurrence > 0:
            entropy -= p_non_occurrence * math.log2(p_non_occurrence)
        # Store the entropy value for the word
        entropy_dict[word] = entropy

    # Sort the words by their entropy (from low to high)
    sorted_entropy = sorted(entropy_dict.items(), key=lambda x: x[1])

    return sorted_entropy

# Example usage
csv_file_path = 'original reviews dataset.csv'  # Update this to your CSV file path
word_list = ['downstairs',
 'dinners',
 'experiences',
 'tv',
 'memorial',
 'outing',
 'floor',
 'grand',
 'refrigerator',
 'excursion',
 'lighting',
 'hyatt',
 'manchester',
 'palace',
 'bedrooms',
 'regency',
 'attending',
 'cleanliness',
 'skyscraper',
 'florist',
 'rooftop',
 'grill',
 'feather',
 'central',
 'indoors',
 'hancok',
 'exploring',
 'seminar',
 'pets',
 'marriott',
 'guestroom',
 'hypo',
 'granddaughters',
 'lamp',
 'queen',
 'riverside',
 'renovations',
 'elegance',
 'continental',
 '173',
 'affina',
 'priceline',
 'views',
 'preferred',
 'ambassador',
 'graduation',
 'bridgestone',
 'towers',
 'sunbathing',
 'welcome',
 'visits',
 'rates',
 'hallways',
 'enjoyable',
 'swissotel',
 'setup',
 'doorman',
 'cubs',
 'barbara',
 'touristy',
 'wastebasket',
 'regal',
 'tour',
 'serene',
 'cities',
 'families',
 'construction',
 'boarding',
 'bedroom',
 'desk',
 'decorating',
 'sundeck',
 'doormen',
 'weekend',
 'spaciousness',
 'riverwalk',
 'hancock',
 'winter',
 'surround',
 'cousin',
 'los',
 'bliss',
 'tower',
 'venues',
 'remembers',
 'jazz',
 'slumber',
 'valets',
 'building',
 'aaa',
 'terrace',
 'penthouse',
 'quilts',
 'accommodation',
 'pantry',
 'staying',
 'bose',
 'millenium',
 'pampered',
 'venue',
 'halls',
 'navy',
 'husband',
 'sofabed',
 'motel',
 'brides',
 'deals',
 'bridemaid',
 'christmas',
 'leisure',
 'luxuries',
 'mojitos',
 'ireland',
 'mile',
 'bars',
 'bagpacks',
 'schmaltzy',
 'valet',
 'parking',
 'walking',
 'cakes',
 'renovation',
 'overnight',
 'showers',
 'suburbs',
 'curtains',
 'upgraded',
 'neck',
 'workstation',
 'bottles',
 'lobby',
 'foodcourt',
 'breathtaking',
 'oatmeal',
 'starwood',
 'facade',
 'westfield',
 'fridge',
 'nightcap',
 'met',
 'store',
 'arrival',
 'traveller',
 'homewoods',
 'hilton',
 'trip',
 'starbucks',
 'vacationed',
 'marble',
 'london',
 'high',
 'michigan',
 'reception',
 'lower',
 'treat',
 'accessible',
 'offices',
 'travelled',
 'view',
 'comfortably',
 'conventions',
 '41st',
 'workout',
 'concierge',
 'upscale',
 'furniture',
 'chechecked',
 'bed',
 'scent',
 'garage',
 'comfortable',
 'movies',
 'roomy',
 'access',
 'hyatts',
 'celebrated',
 'guest',
 'candace',
 'conciergies',
 'aroma',
 'club',
 'cruise',
 'jacuzzi',
 'talbott',
 'paradise',
 'amenities',
 'spacious',
 'tribune',
 'nightly',
 'pillows',
 'internet',
 'rituals',
 'flew',
 'artists',
 'visited',
 'upstairs',
 'considering',
 'concern',
 'renovated',
 'centrally',
 'bucks',
 'tourists',
 'condos',
 'patios',
 'cab',
 'grandaughter',
 'leaving',
 'skyrise',
 'parks',
 'omelet',
 'pier',
 'weekends',
 'coffeepot',
 'recomend',
 'stays',
 'magnificent',
 'westin',
 'train',
 'furnishings',
 'river',
 'buildings',
 'nap',
 'aquarium',
 'roomkeeping',
 'tub',
 'nordstroms',
 'rental',
 'cobb',
 'breakfast',
 'toiletries',
 'seeing',
 'scenery',
 'pillow',
 'attendees',
 'cost',
 'linen',
 'teenage',
 'adequet',
 'maid',
 'tastefully',
 'hospital',
 'attend',
 'garishness',
 'escalator',
 'print',
 'stalls',
 'facet',
 'sincerely',
 'allergies',
 'clubs',
 'ballard',
 'taxis',
 'attractions',
 'swisshotel',
 'knikerbocker',
 'marshall',
 'bathrobes',
 'luxury',
 'coleman',
 'porter',
 'receptionist',
 'vacation',
 'hard',
 'animals',
 'plasma',
 'conrad',
 'travelling',
 'fl',
 'luxuary',
 'station',
 'showplaces',
 'hip',
 'gift',
 'courtesy',
 'macys',
 'sears',
 'layouts',
 'traveling',
 'beds',
 'midwest',
 'cleanest',
 'darkness',
 'rooms',
 'amenity',
 'bathrooms',
 'airlines',
 'airport',
 'annoyed',
 'band',
 'noisy',
 'northbridge',
 'hors',
 'luxurious',
 'european',
 'fireworks',
 'baths',
 'lakeview',
 'nieces',
 'classy',
 'atlanta',
 'showerhead',
 'strudel',
 'el',
 'grandma',
 'husbands',
 'foam',
 'convention',
 'honeymoon',
 'linda',
 'fixtures',
 'affinia',
 'lounge',
 'elegant',
 'supper',
 'resturants',
 'drowning',
 'opinion',
 'soothing',
 'parked',
 'pleasure',
 'immaculate',
 'relaxing',
 'dog',
 'hotwire',
 'groceries',
 'festival',
 'firestone',
 'vouchers',
 'benefits',
 'windy',
 'health',
 'lit',
 'bouquet',
 'stores',
 'situated',
 'remodeled',
 'hotel',
 'ceilings',
 'weeknights',
 'tours',
 'getaway',
 'employees',
 'brightly',
 'designs',
 'tripped',
 'couch',
 'choice',
 'suites',
 'talbot',
 'voucher',
 'paula',
 'air',
 'emergency',
 'modernized',
 'mattress',
 'doggy',
 'nightlife',
 'resort',
 'east',
 'travelzoo',
 'pool',
 'prompt',
 'ballroom',
 'taxi',
 'meeting',
 'visitor',
 'sofa',
 'lodging',
 'july',
 'bus',
 'facilities',
 'holiday',
 'skyline',
 'neatly',
 'bottle',
 'hairdryer',
 'beatles',
 'heat',
 'illinois',
 'geneva',
 'cart',
 'years',
 'steakhouse',
 'arriving',
 'judy',
 'stress',
 'seasons',
 'softest',
 'clientel',
 'texas',
 'pda',
 'superbly',
 'martinis',
 'relaxation',
 'monaco',
 'suggested',
 'remarkable',
 'sheraton',
 'massage',
 'mall',
 'smoking',
 'fitzpatrick',
 'perla',
 'start',
 'ihome',
 'machine',
 'stranded',
 'exhibit',
 'accommodations',
 'museums',
 'booking',
 'courteous',
 'lake',
 'fortney',
 'modern',
 'nights',
 'calories',
 'inn',
 'website',
 'clerks',
 'office',
 'charge',
 'vicinity',
 'spa',
 'center',
 'temp',
 '99',
 'musical',
 'elevator',
 'musically',
 'laptop',
 'sunbeds',
 'spend',
 'bloomingdales',
 'crib',
 'signage',
 'conference',
 'cinema',
 'star',
 'toilet',
 'slept',
 'accomodations',
 '1900',
 'maintained',
 'pleasantly',
 'villa',
 'limo',
 'remote',
 'hotels',
 'fitness',
 'splendido',
 'docking',
 'asthma',
 'membership',
 'house',
 'airline',
 'escape',
 'days',
 'pump',
 'paid',
 'daily',
 'biggest',
 'spotless',
 'towels',
 'palm',
 'tired',
 'maids',
 'hospitality',
 'buffet',
 'exceptional',
 'guestrooms',
 '42nd',
 'ipod',
 'homewood',
 'knickerbocker',
 'hassle',
 'frigid',
 'wireless',
 'institute',
 'explore',
 'lounging',
 'renting',
 'sundecks',
 'rent',
 'travels',
 'mi',
 'motels',
 'rested',
 'walked',
 'housekeeping',
 'married',
 'amazed',
 'buckwheat',
 'cabs',
 'wedding',
 'impersonal',
 'beautiful',
 'sightseeing',
 'reunion',
 'wrigley',
 'covers',
 'restful',
 'massages',
 'rented',
 'noise',
 'memorabilia',
 'amabassador',
 'mojito',
 'relax',
 'gym',
 'hare',
 'redesigned',
 'usa',
 'delighted',
 'roof',
 'entrance',
 'hall',
 'museum',
 'atrium',
 'dec',
 'spent',
 'bellhop',
 'nordstrom',
 'comforting',
 'omni',
 'angeles',
 'shower',
 'definitly',
 'coast',
 'clients',
 'bedding',
 'beach',
 'concourse',
 'indoor',
 'cleaning',
 'suite',
 'kennel',
 'furnished',
 'subway',
 'john',
 'fairmont',
 'rendezvous',
 'vacations',
 'apartments',
 'windchills',
 'pedicure',
 'stay',
 'flowers',
 'pricepoint',
 'fiance',
 'inviting',
 'basement',
 'inns',
 'property',
 'shops',
 'pet',
 'sound',
 'mansions',
 'resorts',
 'buses',
 'complaints',
 'looking',
 'allergy',
 'restaraunts',
 'cta',
 'comfortal',
 'gorgeous',
 'wi',
 'osco',
 'alarm',
 'bath',
 'exceptionally',
 'overpriced',
 'spectacular',
 'celebration',
 'cabinet',
 'bugs',
 'sun',
 'cheaper',
 'anniversary',
 'clerk',
 'sauna',
 'flight',
 'halibut',
 'cornell',
 'couples',
 'millennium',
 'feathers',
 'rate']

# Calculate and sort the entropy
sorted_entropy = calculate_entropy_from_csv_pandas(csv_file_path, word_list)


print("Sorted entropy (low to high):")
for word, entropy in sorted_entropy:
    print(f"{word}: {entropy}")
```
